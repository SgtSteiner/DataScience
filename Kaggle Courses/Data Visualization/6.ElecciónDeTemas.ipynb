## Elección de tipos de trama y estilos personalizados

En este micro curso, aprendimos a crear muchos tipos de gráficos diferentes. Ahora, organizarás tu conocimiento, antes de aprender algunos comandos rápidos que puedes usar para cambiar el estilo de tus gráficos.

![¿Que has aprendido?](https://imgur.com/2VmgDnF.png)

Dado que no siempre es fácil decidir cómo contar mejor la historia detrás de tus datos, hemos dividido los tipos de gráficos en tres categorías generales para ayudar con esto.

+ **Tendencias** - una tendencia se define como un patrón de cambio.
  + `sns.lineplot` - los **gráficos de líneas** son mejores para mostrar tendencias durante un período de tiempo y se pueden usar varias líneas para mostrar tendencias en más de un grupo.


+ **Relación** - existen muchos tipos de gráficos diferentes que podemos usar para comprender las relaciones entre las variables en los datos.
  + `sns.barplot` - los **gráficos de barras** son útiles para comparar cantidades correspondientes a diferentes grupos.
  + `sns.heatmap` - los **mapas de calor** se pueden usar para buscar patrones codificados por colores en tablas de números.
  + `sns.scatterplot` - los **diagramas de dispersión** muestran la relación entre dos variables continuas; si está codificado por colores, también podemos mostrar la relación con una tercera variable categórica.
  + `sns.regplot` - incluir una **línea de regresión** en el diagrama de dispersión facilita ver cualquier relación lineal entre dos variables.
  + `sns.lmplot` - este comando es útil para dibujar varias líneas de regresión, si el diagrama de dispersión contiene múltiples grupos codificados por colores.
  + `sns.swarmplot`: los **diagramas de dispersión categórica** muestran la relación entre una variable continua y una variable categórica.


+ **Distribución**: visualizamos distribuciones para mostrar los posibles valores que podemos esperar ver en una variable, junto con la probabilidad de que sean.
  + `sns.distplot` - los **histogramas** muestran la distribución de una sola variable numérica.
  + `sns.kdeplot` - los **gráficos KDE** (o **gráficos 2D KDE**) muestran una distribución estimada y uniforme de una sola variable numérica (o dos variables numéricas).
  + `sns.jointplot` - este comando es útil para mostrar simultáneamente un diagrama de KDE 2D con los diagramas de KDE correspondientes para cada variable individual.

### Cambiando estilos con seaborn

Todos los comandos han proporcionado un bonito estilo predeterminado a cada una de los gráficos. Sin embargo, puede resultar útil personalizar el aspecto de los gráficos y, afortunadamente, esto se puede lograr simplemente agregando una línea más de código.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Ruta del archivo a leer
spotify_filepath = "./input/spotify.csv"

# Lee el archivo en la variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)

# Gráfico de líneas
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data);
```

Podemos cambiar rápidamente el estilo de la figura a un tema diferente con una sola línea de código.

```python
# Cambia el estilo de la figura al tema "dark"
sns.set_style("dark")

# Gráfico de líneas
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data);
```

Seaborn tiene cinco temas diferentes: (1) `"darkgrid"`, (2) `"whitegrid"`, (3) `"dark"`, (4) `"white"` y (5) `"ticks"`, y solo necesitas usar un comando similar al que está en la celda de código anterior (con el tema elegido completado) para cambiarlo.
