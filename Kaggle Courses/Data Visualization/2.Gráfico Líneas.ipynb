### Dibujar los datos

¡Ahora que el conjunto de datos está cargado en el notebook, solo necesitamos una línea de código para hacer un gráfico de líneas!

```python
# Gráfico de líneas mostrando el consumo global diario de cada canción
sns.lineplot(data=spotify_data);
```

Como puede verse arriba, la línea de código es relativamente corta y tiene dos componentes principales:

+ `sns.lineplot` le dice al notebook que queremos crear un gráfico de líneas.
  + Cada comando que conozca en este curso comenzará con **`sns`**, lo que indica que el comando proviene del paquete [seaborn](https://seaborn.pydata.org/). Por ejemplo, usamos **`sns.lineplot`** para hacer gráficos de líneas. Pronto, aprenderá que usamos **`sns.barplot`** y **`sns.heatmap`** para hacer gráficos de barras y mapas de calor, respectivamente.
+ `data = spotify_data` selecciona los datos que se utilizarán para crear el gráfico.

Tenga en cuenta que siempre usará este mismo formato cuando cree un gráfico de líneas y lo único que cambia con un nuevo conjunto de datos es el nombre del conjunto de datos. Entonces, si estaba trabajando con un conjunto de datos diferente llamado `financial_data`, por ejemplo, la línea de código aparecería de la siguiente manera:

`sns.lineplot(data=financial_data)`

A veces hay detalles adicionales que nos gustaría modificar, como el tamaño de la figura y el título del gráfico. Cada una de estas opciones se puede configurar fácilmente con una sola línea de código.

```python
# Establece el ancho y la altura de la figura
plt.figure(figsize=(14,6))

# Añade título
plt.title("Consumo Global Diario de Canciones Populares en 2017-2018")

# Gráfico de líneas mostrando el consumo global diario de cada canción
sns.lineplot(data=spotify_data);
```

La primera línea de código establece el tamaño de la figura en `14` pulgadas (en ancho) por `6` pulgadas (en altura). Para establecer el tamaño de cualquier figura, solo necesita copiar la misma línea de código que aparece. Luego, si desea utilizar un tamaño personalizado, cambie los valores proporcionados de 14 y 6 al ancho y alto deseados.

La segunda línea de código establece el título de la figura. ¡Tenga en cuenta que el título siempre debe estar entre comillas `("...")`!

### Dibujar un subconjunto de datos

Hasta ahora, ha aprendido a dibujar una línea para cada columna del conjunto de datos. En esta sección, aprenderá a dibujar un subconjunto de las columnas.

Comenzaremos imprimiendo los nombres de todas las columnas. Esto se hace con una línea de código y se puede adaptar para cualquier conjunto de datos simplemente intercambiando el nombre del conjunto de datos (en este caso, `spotify_data`).

```python
spotify_data.columns
```

En la siguiente celda de código dibujamos las líneas correspondientes a las dos primeras columnas del conjunto de datos.

```python
# Establece el ancho y la altura de la figura
plt.figure(figsize=(14,6))

# Añade título
plt.title("Consumo Global Diario de Canciones Populares en 2017-2018")

# Gráfico de línea mostrando el consumo global diario de 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

# Gráfico de línea mostrando el consumo global diario de 'Despacito'
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")

# Añade etiqueta para el eje horizontal
plt.xlabel("Date");
```

Las primeras dos líneas de código establecen el título y el tamaño de la figura. Las siguientes dos líneas agregan una línea al gráfico de líneas. Por ejemplo, considere la primera que agrega la línea para "Shape of You":

`sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")`

Esta línea se ve muy similar al código que usamos cuando dibujamos cada línea en el conjunto de datos, pero tiene algunas diferencias clave:

+ En lugar de establecer `data = spotify_data`, establecemos `data = spotify_data ['Shape of You']`. En general, para dibujar solo una columna, usamos este formato poniendo el nombre de la columna entre comillas simples y encerrándolo entre corchetes. (*Para asegurarse de especificar correctamente el nombre de la columna, puede imprimir la lista de todos los nombres de columna con el comando que vimos anteriormente*).
+ También agregamos `label = "Shape of You"` para que la línea aparezca en la leyenda y establezca su etiqueta correspondiente.

La línea final de código modifica la etiqueta para el eje horizontal (o eje x), donde la etiqueta deseada se coloca entre comillas `("...")`.
