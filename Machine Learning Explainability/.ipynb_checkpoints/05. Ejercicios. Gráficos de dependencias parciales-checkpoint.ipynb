## Configuración

Vamos a crear gráficos de dependencias parciales y practicar la construcción de insights con datos de la competición de [Preedicción de tarifas de taxi](https://www.kaggle.com/c/new-york-city-taxi-fare-prediction).

Te hemos proporcionado otra vez el código básico de carga, revisión y construcción del modelo. Ejecuta la celda siguiente para configurarlo todo:

````python
# Carga de datos, división, modelado y EDA
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Data manipulation code below here
data = pd.read_csv('./data/new-york-city-taxi-fare-prediction/train.csv', nrows=50000)

# Eliminamos datos de coordenadas con valores atípicos extremos o tarifas negativas
data = data.query('pickup_latitude > 40.7 and pickup_latitude < 40.8 and ' +
                  'dropoff_latitude > 40.7 and dropoff_latitude < 40.8 and ' +
                  'pickup_longitude > -74 and pickup_longitude < -73.9 and ' +
                  'dropoff_longitude > -74 and dropoff_longitude < -73.9 and ' +
                  'fare_amount > 0'
                  )

y = data.fare_amount

base_features = ['pickup_longitude',
                 'pickup_latitude',
                 'dropoff_longitude',
                 'dropoff_latitude']

X = data[base_features]


train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
first_model = RandomForestRegressor(n_estimators=30, random_state=1).fit(train_X, train_y)

print("Data sample:")
data.head()
````

````python
data.describe()
````

## Pregunta 1

Aquí está el código para crear el gráfico de dependencias parciales para la longitud de recogida. Ejecuta la siguiente celda.

````python
from matplotlib import pyplot as plt
from pdpbox import pdp, get_dataset, info_plots

%matplotlib inline

feat_name = 'pickup_longitude'
pdp_dist = pdp.pdp_isolate(model=first_model, dataset=val_X, model_features=base_features, feature=feat_name)

pdp.pdp_plot(pdp_dist, feat_name)
plt.show()
````

¿Por qué la gráfica de dependencia parcial tiene esta forma de U?

¿Su explicación sugiere qué forma esperar en las gráficas de dependencia parcial para las otras características?

Cree todos los demás gráficos parciales en un bucle-for a continuación (copiando las líneas correspondientes del código anterior).

````python
for feat_name in base_features:
    pdp_dist = pdp.pdp_isolate(model=first_model, dataset=val_X, model_features=base_features, feature=feat_name)

    pdp.pdp_plot(pdp_dist, feat_name)
    plt.show()
````

¿Las formas se ajustan a sus expectativas sobre qué formas tendrían? ¿Puedes explicar la forma ahora que los has visto?

Descomenta la siguiente línea para comprobar tu intuición.

***Solución***: Desde la importancia de la permutación, tenemos la sensación de que la distancia es el determinante más importante de la tarifa del taxi.

Este modelo no incluyó medidas de distancia (como el cambio absoluto en latitud o longitud) como características, por lo que las características de coordenadas (como pickup_longitude) capturan el efecto de la distancia. El hecho de ser recogido cerca del centro de los valores de longitud reduce las tarifas pronosticadas en promedio, porque significa viajes más cortos (en promedio).

Por la misma razón, vemos la forma general de U en todas nuestras gráficas de dependencia parcial.

## Pregunta 4

En los PDP que has visto hasta ahora, las funciones de ubicación han servido principalmente como un proxy para capturar la distancia recorrida. En las lecciones de importancia de permutación, agregaste las características `abs_lon_change` y` abs_lat_change` como una medida más directa de la distancia.

Crea estas características de nuevo aquí. Solo necesitas completar las dos líneas superiores. A continuación, ejecuta la siguiente celda.

**Después de ejecutarlo, identifica la diferencia más importante entre esta gráfica de dependencia parcial y la que obtuviste sin características de valor absoluto. El código para generar el PDP sin características de valor absoluto se encuentra en la parte superior de esta celda de código.**

---

````python
# Este es el PDP para pickup_longitude sin las características de diferencias absolutas. Se incluye aquí para compararla con la nueva PDP a crear
feat_name = 'pickup_longitude'
pdp_dist_original = pdp.pdp_isolate(model=first_model, dataset=val_X, model_features=base_features, feature=feat_name)

pdp.pdp_plot(pdp_dist_original, feat_name)
plt.show()



# Se crean las nuevas características
data['abs_lon_change'] = abs(data.dropoff_longitude - data.pickup_longitude)
data['abs_lat_change'] = abs(data.dropoff_latitude - data.pickup_latitude)

features_2  = ['pickup_longitude',
               'pickup_latitude',
               'dropoff_longitude',
               'dropoff_latitude',
               'abs_lat_change',
               'abs_lon_change']

X = data[features_2]
new_train_X, new_val_X, new_train_y, new_val_y = train_test_split(X, y, random_state=1)
second_model = RandomForestRegressor(n_estimators=30, random_state=1).fit(new_train_X, new_train_y)

feat_name = 'pickup_longitude'
pdp_dist = pdp.pdp_isolate(model=second_model, dataset=new_val_X, model_features=features_2, feature=feat_name)

pdp.pdp_plot(pdp_dist, feat_name)
plt.show()
````

***Solución***: La mayor diferencia es que el gráfico de dependencia parcial se hizo mucho más pequeño. El valor vertical más bajo es aproximadamente 15 por debajo del valor vertical más alto en el gráfico superior, mientras que esta diferencia es solo de 3 en el gráfico que acabas de crear. En otras palabras, una vez que controlas la distancia absoluta recorrida, pickup_longitude tiene un impacto muy pequeño en las predicciones.

## Pregunta 6
El siguiente código hace lo siguiente:

1. Crea dos características, `X1` y `X2`, que tienen valores aleatorios en el rango [-2, 2].
2. Crea una variable objetivo `y`, que siempre es 1.
3. Entrena un modelo `RandomForestRegressor` para predecir `y` dados `X1` y `X2`.
4. Crea una gráfica PDP para `X1` y una gráfica de dispersión de `X1` frente a `y`.

¿Tienes una predicción acerca de cómo se verá la gráfica PDP? Ejecuta la celda para averiguarlo.

Modifique la inicialización de `y` para que nuestra gráfica PDP tenga una pendiente positiva en el rango [-1,1], y una pendiente negativa en todas partes. (Nota: *solo debes modificar la creación de `y`, sin modificar `X1`, `X2` y `my_model`.*)

````python
import numpy as np
from numpy.random import rand

n_samples = 20000

# Create array holding predictive feature
X1 = 4 * rand(n_samples) - 2
X2 = 4 * rand(n_samples) - 2
# Create y. you should have X1 and X2 in the expression for y
# y = np.ones(n_samples) 
y = -2 * X1 * (X1<-1) + X1 - 2 * X1 * (X1>1) - X2

# create dataframe because pdp_isolate expects a dataFrame as an argument
my_df = pd.DataFrame({'X1': X1, 'X2': X2, 'y': y})
predictors_df = my_df.drop(['y'], axis=1)

my_model = RandomForestRegressor(n_estimators=30, random_state=1).fit(predictors_df, my_df.y)

pdp_dist = pdp.pdp_isolate(model=my_model, dataset=my_df, model_features=['X1', 'X2'], feature='X1')

# visualize your results
pdp.pdp_plot(pdp_dist, 'X1')
plt.show()
````

## Pregunta 7

Cree un conjunto de datos con 2 funciones y un objetivo, de manera que el pdp de la primera función sea plano, pero su importancia de permutación sea alta. Usaremos un RandomForest para el modelo.

*Nota: Solo necesita suministrar las líneas que crean las variables X1, X2 y y. Se proporciona el código para construir el modelo y calcular las percepciones*.

````python
import eli5
from eli5.sklearn import PermutationImportance

n_samples = 20000

# Create array holding predictive feature
X1 = 4 * rand(n_samples) - 2
X2 = 4 * rand(n_samples) - 2
# Create y. you should have X1 and X2 in the expression for y
y = X1 * X2

# create dataframe because pdp_isolate expects a dataFrame as an argument
my_df = pd.DataFrame({'X1': X1, 'X2': X2, 'y': y})
predictors_df = my_df.drop(['y'], axis=1)

my_model = RandomForestRegressor(n_estimators=30, random_state=1).fit(predictors_df, my_df.y)


pdp_dist = pdp.pdp_isolate(model=my_model, dataset=my_df, model_features=['X1', 'X2'], feature='X1')
pdp.pdp_plot(pdp_dist, 'X1')
plt.show()

perm = PermutationImportance(my_model).fit(predictors_df, my_df.y)

# show the weights for the permutation importance you just calculated
eli5.show_weights(perm, feature_names = ['X1', 'X2'])
````
